```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV
df = pd.read_csv("./dash/new.csv")
each_data = df["SOURCE_SYSTEM"].unique()

# Dictionary to store the datasets
data_set_lists = {}
for i in each_data:
    filtered_df = df.loc[df["SOURCE_SYSTEM"] == i, ["RECORD_DATE", "TOTAL MANUAL EFFORTS"]]
    data_set_lists[i] = filtered_df  # Store filtered data
plt.figure(figsize=(16, 6))
for key, data in data_set_lists.items():
    plt.plot(data["RECORD_DATE"], data["TOTAL MANUAL EFFORTS"], label=key, marker='o')  # Added dots    
    for x, y in zip(data["RECORD_DATE"], data["TOTAL MANUAL EFFORTS"]):
            plt.text(x, y, str(y), fontsize=9, ha="right", va="bottom")

# Customize plot
plt.xlabel("Record Date")
plt.ylabel("Total Manual Efforts")
plt.title("Total Manual Efforts Over Time by Source System")
plt.legend()  # Add legend to differentiate lines
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid

# Reverse the x-axis to show the most recent date on the left
plt.gca().invert_xaxis()

image_path = "./dash/manual_efforts_chart.png"
plt.savefig(image_path, bbox_inches="tight")  # Save with tight layout

# Show the plot
plt.show()
plt.close() 

```
